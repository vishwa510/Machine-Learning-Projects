# Data Preprocessing - Separating features and target variable

```python
X = df.drop(labels = ['tpep_dropoff_datetime', 'tpep_pickup_datetime', 'total_amount'], axis=1)
y = df['total_amount']
```


# Feature Engineering - Encoding categorical features

```python
X1=pd.get_dummies(X,columns=['store_and_fwd_flag',
                             'payment_type'],drop_first='True')
```

```python
X1.columns
```

# Detecting missing data and imputation

```python
from sklearn.impute import SimpleImputer
imp_median = SimpleImputer(missing_values=np.nan, strategy='median')
X1 = imp_median.fit_transform(X1)
# X1 = pd.DataFrame(X1, columns=[
#         'VendorID', 'passenger_count', 'trip_distance', 'RatecodeID',
#        'store_and_fwd_flag', 'PULocationID', 'DOLocationID', 'payment_type',
#        'extra', 'tip_amount', 'tolls_amount', 'improvement_surcharge',
#        'congestion_surcharge', 'Airport_fee'])
```


# Splitting the data into training and testing sets

```python
from sklearn.model_selection import train_test_split

X1_train, X1_test, y_train, y_test = train_test_split(
    X1, y, test_size=0.2, random_state=42)
```


# Training and evaluating Linear Regression model

#  #hyperparameter tuning not needed used default hyperparameters

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr = lr.fit(X1_train, y_train)
lr.score(X1_test, y_test)
```


# Training and evaluating Decision Tree model

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor(random_state=0)
dtr = dtr.fit(X1_train, y_train)
dtr.score(X1_test, y_test)
```

# Training and evaluating Random Forest model

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(random_state=0)
rfr = rfr.fit(X1_train, y_train)
rfr.score(X1_test, y_test)
```


# Loading test dataset for submission

```python
X2 = pd.read_csv('/kaggle/input/taxi-fare-guru-total-amount-prediction-challenge/test.csv')
```

```python
X2 = X2.drop(labels = ['tpep_dropoff_datetime', 'tpep_pickup_datetime'], axis=1)
```

```python
X3=pd.get_dummies(X2,columns=['store_and_fwd_flag',
                             'payment_type'],drop_first='True')
```

```python
X3.columns
```


# Imputing missing values in the test dataset

```python
from sklearn.impute import SimpleImputer
imp_median = SimpleImputer(missing_values=np.nan, strategy='median')
X3 = imp_median.fit_transform(X3)
```

# Generating predictions for submission using the best model (Random Forest)

```python
submission = rfr.predict(X3)
```

```python
submission = pd.DataFrame(submission, columns=['total_amount'])
```

```python
submission.index.name = 'ID'
submission.index = submission.index + 1
```
